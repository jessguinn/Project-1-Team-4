```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Import Data
df1 = pd.read_csv("/Users/makenzieeason/Desktop/Gender_Diabetes_2000-2021.csv")

# Display column names
column_names = df1.columns
print(column_names)
```

```python
# Convert 'PercentageFemale' to numeric and drop rows with missing or non-numeric values
df1['PercentageFemale'] = pd.to_numeric(df1['PercentageFemale'], errors='coerce')
df1 = df1.dropna(subset=['PercentageFemale'])

# Convert 'PercentageMale' to numeric and drop rows with missing or non-numeric values
df1['PercentageMale'] = pd.to_numeric(df1['PercentageMale'], errors='coerce')
df1 = df1.dropna(subset=['PercentageMale'])

# Find the top 10 states with the highest percentages for females
top_female_states = df1.groupby('StateFemale').mean().nlargest(10, 'PercentageFemale').index
top_female = df1[df1['StateFemale'].isin(top_female_states)]

# Find the top 10 states with the highest percentages for males
top_male_states = df1.groupby('StateMale').mean().nlargest(10, 'PercentageMale').index
top_male = df1[df1['StateMale'].isin(top_male_states)]

# Set style for better readability
sns.set(style="whitegrid")

# Plot bar plot for the top 10 states with the highest percentages for females
plt.figure(figsize=(14, 7))
sns.barplot(x='StateFemale', y='PercentageFemale', hue='YearFemale', data=top_female, palette='viridis')
plt.title('Top 10 States with Highest Percentage of Females with Diabetes')
plt.xlabel('State')
plt.ylabel('Percentage')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Plot bar plot for the top 10 states with the highest percentages for males
plt.figure(figsize=(14, 7))
sns.barplot(x='StateMale', y='PercentageMale', hue='YearMale', data=top_male, palette='magma')
plt.title('Top 10 States with Highest Percentage of Males with Diabetes')
plt.xlabel('State')
plt.ylabel('Percentage')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Convert 'PercentageFemale' to numeric and drop rows with missing or non-numeric values
df1['PercentageFemale'] = pd.to_numeric(df1['PercentageFemale'], errors='coerce')
df1 = df1.dropna(subset=['PercentageFemale'])

# Convert 'PercentageMale' to numeric and drop rows with missing or non-numeric values
df1['PercentageMale'] = pd.to_numeric(df1['PercentageMale'], errors='coerce')
df1 = df1.dropna(subset=['PercentageMale'])

# Find the top 10 states with the lowest percentages for females
bottom_female_states = df1.groupby('StateFemale').mean().nsmallest(10, 'PercentageFemale').index
bottom_female = df1[df1['StateFemale'].isin(bottom_female_states)]

# Find the top 10 states with the lowest percentages for males
bottom_male_states = df1.groupby('StateMale').mean().nsmallest(10, 'PercentageMale').index
bottom_male = df1[df1['StateMale'].isin(bottom_male_states)]

# Set style for better readability
sns.set(style="whitegrid")

# Plot bar plot for the top 10 states with the lowest percentages for females
plt.figure(figsize=(14, 7))
sns.barplot(x='StateFemale', y='PercentageFemale', hue='YearFemale', data=bottom_female, palette='viridis')
plt.title('Top 10 States with Lowest Percentage of Females with Diabetes')
plt.xlabel('State')
plt.ylabel('Percentage')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Plot bar plot for the top 10 states with the lowest percentages for males
plt.figure(figsize=(14, 7))
sns.barplot(x='StateMale', y='PercentageMale', hue='YearMale', data=bottom_male, palette='magma')
plt.title('Top 10 States with Lowest Percentage of Males with Diabetes')
plt.xlabel('State')
plt.ylabel('Percentage')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
