```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Import Data
df1 = pd.read_csv("/Users/makenzieeason/Desktop/states_merged.csv")

# Display column names
column_names = df1.columns
print(column_names)
```

```python
# Select specific years
selected_years = [2001, 2005, 2011, 2016, 2021]

# Filter DataFrame for selected years
df_selected_years = df1[df1['Year'].isin(selected_years)]

# Grouped bar chart with specific colors
fig, ax = plt.subplots(figsize=(10, 6))

width = 0.2  # Width of the bars
age_groups = df1.columns[2:]
colors = {'18-44 percentage': 'blue', '45-64 percentage': 'green', '65-74 percentage': 'orange', '75+ percentage': 'red'}

for i, year in enumerate(selected_years):
    for j, age_group in enumerate(age_groups):
        values = df_selected_years[df_selected_years['Year'] == year][age_group]
        x_pos = i + j * width

        ax.bar(x_pos, values, width=width, label=f"{year} - {age_group}", color=colors[age_group])

ax.set_xticks([i + width for i in range(len(selected_years))])
ax.set_xticklabels(selected_years)
ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.set_title('Age Group Distribution Across States for Specific Years')

# Move legend outside the box
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

```python
# Select specific years
selected_years = [2001, 2005, 2011, 2016, 2021]

# Filter out 'Median of States'
df_filtered = df1[df1['State'] != 'Median of States']

# Identify top 5 states based on median '18-44 percentage'
top_states_18_44 = df_filtered.groupby('State')['18-44 percentage'].median().nlargest(5).index

# Create subplots for each state
fig, ax = plt.subplots(figsize=(10, 6))

# Iterate over top 5 states
for state in top_states_18_44:
    df_state = df_filtered[df_filtered['State'] == state]
    
    ax.plot(df_state['Year'], df_state['18-44 percentage'], label=state, marker='o')

ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.set_title('Top 5 States - 18-44 Percentage Trend Over the Years')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
