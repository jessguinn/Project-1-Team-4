```python
# Read in the CDC Diagnosed data
diagnosed_df = pd.read_csv('CDC Diagnosed.csv')
diagnosed_df.head()
```

```python
diagnosed_df[["Year", "State", "Diagnosed Diabetes (Percentage) ", "Obesity (Percentage) ", "Newly Diagnosed Diabetes (Rate per 1000)", "Newly Diagnosed Physical Inactivity (Percentage) "]].head()
```

```python
average_obesity = diagnosed_df["Obesity (Percentage) "].mean()
average_obesity
```

```python
average_inactivity = diagnosed_df["Newly Diagnosed Physical Inactivity (Percentage) "].mean()
inactivity = "Newly Diagnosed Diabetes (Rate per 1000)"
inactivity_min = diagnosed_df["Newly Diagnosed Physical Inactivity (Percentage) "].min()
inactivity_max = diagnosed_df["Newly Diagnosed Physical Inactivity (Percentage) "].max()

inactivity_range = inactivity_max - inactivity_min
print(f"The range of values in '{inactivity}' is {inactivity_range}")

inactivity_group = diagnosed_df.groupby(["Year", "State"])["Newly Diagnosed Physical Inactivity (Percentage) "].agg([min, max])
inactivity_group["Range"] = inactivity_group["max"] - inactivity_group["min"]
print(inactivity_group)

average_inactivity
```

```python
inactivity_year_diff = diagnosed_df.groupby("State")[["Year", "Newly Diagnosed Physical Inactivity (Percentage) "]].agg({"Year": ["min", "max"], "Newly Diagnosed Physical Inactivity (Percentage) ": ["min", "max"]})

inactivity_year_diff["Range"] = inactivity_year_diff[("Newly Diagnosed Physical Inactivity (Percentage) ", "max")] - inactivity_year_diff[("Newly Diagnosed Physical Inactivity (Percentage) ", "min")]
inactivity_year_diff["Change"] = inactivity_year_diff[("Newly Diagnosed Physical Inactivity (Percentage) ", "max")] - inactivity_year_diff[("Newly Diagnosed Physical Inactivity (Percentage) ", "min")]

print(inactivity_year_diff)
```
