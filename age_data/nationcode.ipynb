```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
national_age_path = "age_national.csv"

# Read the CSV file and select the desired columns
national_age = pd.read_csv(national_age_path, skiprows=2, usecols=['Year','18-44 - Percentage','45-64 - Percentage','65-74 - Percentage','75+ - Percentage'])

# Rename the columns
national_age = national_age.rename(columns={'18-44 - Percentage': '18-44 percentage',
                                            '45-64 - Percentage': '45-64 percentage',
                                            '65-74 - Percentage': '65-74 percentage',
                                            '75+ - Percentage': '75+ percentage'})

# Select only the desired columns
national_age = national_age[['Year', '18-44 percentage', '45-64 percentage', '65-74 percentage', '75+ percentage']]

# Print the resulting DataFrame
national_age.to_csv('national_age.csv', index=False)



```
